```python
import folium

# Create a Folium map centered on a specific location
map_center = [0.0, 0.0]  # Replace with your desired center coordinates
mymap = folium.Map(location=map_center, zoom_start=12)

# List to store marker coordinates
marker_coordinates = []

# Function to add a marker to the map and store coordinates
def add_marker(map_object, location, popup_text):
    folium.Marker(location, popup=popup_text).add_to(map_object)
    marker_coordinates.append(location)

# Example markers
marker_locations = [(12, 12, 'Marker 1'), (13, 13, 'Marker 2')]

# Add markers to the map
for location in marker_locations:
    add_marker(mymap, (location[0], location[1]), location[2])

# Create a custom JavaScript function to handle click events
js_click_handler = """
function onMapClick(e) {
    var lat = e.latlng.lat;
    var lon = e.latlng.lng;
    L.marker([lat, lon], {popup: 'New Marker'}).addTo(mymap);
    marker_coordinates.push([lat, lon]);
}

mymap.on('click', onMapClick);
"""

# Add the JavaScript click handler to the map
folium.Marker([0, 0], icon=folium.Icon(color='white')).add_to(mymap)  # Add a dummy marker to initialize the map
mymap.get_root().header.add_child(folium.Element("<script>{}</script>".format(js_click_handler)))

# Display the map
mymap
```
